PARTE 1 - m2 variando entre 1 e 10 kg

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros fixos
m1 = 1.0  # massa 1 (kg)
k = 1.0   # constante da mola (N/m)

# Intervalo de m2
m2_values = np.linspace(1, 10, 10)
freqs1 = []
freqs2 = []
eigvecs_list = []  # Lista para armazenar os autovetores

for m2 in m2_values:
    M = np.array([[m1, 0],   # Matriz de massa
                  [0, m2]])
    K = np.array([[2*k, -k],  # Matriz de rigidez
                  [-k, 2*k]])
    eigvals, eigvecs = np.linalg.eig(np.linalg.inv(M) @ K)  # Calcula os autovalores e autovetores
    idx = np.argsort(eigvals)  # Ordena os autovalores
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:, idx]

    freqs = np.sqrt(eigvals)  # Frequências naturais (rad/s)
    # Garante que a primeira frequência seja exatamente zero (modo de translação)
    freqs[0] = 0.0
    
    freqs1.append(freqs[0])
    freqs2.append(freqs[1])
    eigvecs_list.append(eigvecs)  # Armazena os autovetores

plt.scatter(m2_values, freqs1, label='Frequência 1 (modo de translação)', color='blue', s=15)
plt.scatter(m2_values, freqs2, label='Frequência 2 (modo relativo)', color='red', s=15)
plt.xlabel('m2 (kg)')
plt.ylabel('Frequência normal (rad/s)')
plt.title('Frequências normais vs m2 (sistema cíclico)')
plt.legend()
plt.grid(True)
plt.show()

# Exemplo: imprimir autovetores para todos os valores de m2
for i, m2 in enumerate(m2_values):
    print(f"Autovetores para m2 = {m2:.1f} kg:")
    print(eigvecs_list[i])
    print("-" * 30)
```
